```python
N,K = map(int,input().split())
arr = [i for i in range(1,N+1)]    # 맨 처음에 원에 앉아있는 사람들

answer = []                         # 제거된 사람들 리스트
num = 0                          

for t in range(N):
    num += K-1  
    if num >= len(arr):             # 한바퀴를 돌고 그 다음으로 돌아올때  
        num = num%len(arr)
 
    answer.append(str(arr.pop(num)))
print("<",", ".join(answer)[:],">", sep='')
```
