```python
C = int(input())

for i in range(C):
  nums = list(map(int, input().split()))  # 숫자를 입력받고 리스트화
  avg = sum(nums[1:])/nums[0]             # 점수들의 합 / 학생 수
  h_avg = 0                               # 초기 평균보다 높은 학생 수 설정
  for score in nums[1:]:                  # 점수 안에서 for문
    if score > avg:                       # 평균보다 높은 점수 조건 if 문
      h_avg += 1                          # 학생수 +1 설정
  avg_rate = h_avg/nums[0] *100           # 평균보다 높은 학생 비율 구하기
  print(f'{avg_rate:.3f}%')               # 소수점 셋째자리까지 설정
```
