```python
T = int(input())
for i in range(T):
  H,W,N = map(int, input().split())
  a = N % H                         # 나머지를 통해 라인 배정
  b = N//H+1                        # 호수 배정
  if a == 0:                        # 전체 높이와 맞아 떨어지는 손님
    a = H                           
    b -= 1                         
  print(a*100+b)
```
