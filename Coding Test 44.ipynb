```python
word = input()
alpabet = ['ABC','DEF','GHI','JKL','MNO','PQRS','TUV','WXYZ']
time=0
for dial in alpabet :  
    for i in dial:  # alpabet 에서 각 요소를 꺼내서 한글자씩 분리
        for x in word :  # 입력받은 문자를 하나씩 분리
            if i == x :  
                time += alpabet.index(dial) +3  # time = time + index +3
print(time)
```
