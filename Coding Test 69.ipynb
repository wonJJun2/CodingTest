```python
def draw_stars(n):
    matrix = []
    for i in range(3 * len(n)):
        if i // len(n) == 1:
            matrix.append(n[i % len(n)] + " " * len(n) + n[i % len(n)])
        else:
            matrix.append(n[i % len(n)] * 3)
    return matrix
 
 
star = ["***", "* *", "***"]
n = int(input())
e = 0
while n != 3:
    n = int(n / 3)
    e += 1
 
for i in range(e):
    star = draw_stars(star)
for i in star:
    print(i)

# 다른사람의 숏코딩
n=int(input())
s='*'
while n>1:
 t=[i*3 for i in s]
 s=t+[i+' '*len(i)+i for i in s]+t
 n//=3
print('\n'.join(s))
```
