```python
N = int(input())
group_word = 0
for i in range(N):
  word = input()
  not_group = 0
  for j in range(len(word)-1):
    if word[j] != word[j+1]:          # 뒤의 단어가 다른 단어일 때
      new_word = word[j+1:]           # 새로운 단어로 인식
      if new_word.count(word[j]) > 0: # 남은 문자열에서 해당 글자가 있으면
        not_group += 1                # 있으면 그룹단어가 아니다
  if not_group == 0:                  # 남은 문자열에서 해당 단어가 없으면
    group_word += 1                   # 그룹 단어에 추가
print(group_word)
```
