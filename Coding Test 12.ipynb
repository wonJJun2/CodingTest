```python
H,M = map(int,input().split())

if M <= 44:         # 분이 44분보다 작거나 같을때         
  if H == 0 :       # 이중 조건문, H == 0 이면,
    H = 23          # H = 23 입력 (시간은 0에서 23 사이 값을 가진다.)
    M += 60         # M 입력 값에 60을 더해서 입력한다.
  else :            # H == 0 이 아니면
    H -= 1          # H에서 1을 뺀값을 입력한다.
    M += 60         # M 입력 값에 60을 더해서 입력한다.

print(H,M-45)       # 45분 일찍 일어나기 위해 45를 빼고 출력한다.
```
